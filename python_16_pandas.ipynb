iloc는 indexer를 이용해서 가져오고, loc는 이름(컬럼명)을 이용해서 가져온다.

```python
# 행은 이름이 설정되어 있지 않아서 자동으로 생성된 Range Index가 이름이 된다.
data.loc[0, 'tip']
```

```python
data.loc[0:5, 'tip':'time']
```

```python
data.loc[[0, 3, 5], ['tip', 'day', 'time']]
```

### at
이름으로 한 위치의 값만 가져온다.

```python
data.at[0, 'tip']
# data.loc[0, 'tip']
# data.iloc[0, 1]
```

```python
# at은 slicing을 사용할 수 없다.
data.at[0:3, 'tip']
```

```python
# at은 fancy indexing을 사용할 수 없다
data.at[[1, 3, 5], ['tip', 'smoker', 'day']]
```

### iat
iat은 index 번호로 한 위치의 값을 가져온다.  
integer, indexers 만 사용할 수 있다.

```python
# at은 loc로, iat은 index로
data.iat[0, 1]
# data.at[0, 'tip']
```

```python
# iat은 slicing을 사용할 수 없다.
# iat은 fancy indexing을 사용할 수 없다.
data.iat[0:3, 1]
```

### Boolean Indexing

```python
# data.tip >= 5
data['tip'] >= 5
```

```python
# tip값이 5 이상인것만 출력
data[data['tip'] >= 5] # data[True, False, True, True....]
```

```python
data.tip
data['tip']

# data[data['tip'] >= 5]
# a['tip']

print(data[data['tip'] >= 5]['tip']) # data[data['tip'] >= 5] 의 결과는 DataFrame
data['tip'][3] # data['tip]의 결과는 Series
```

```python
data[data['tip'] >= 5].loc[:, 'tip':'day']
```

```python
data[data['tip'] >= 5].iloc[:, 1:5]
```

```python
data[data['tip'] >= 5][['tip','day']]
```

요일이 'Sun'이거나 'Sat'의 tip과 day의 값 가져오기

```python
data[(data['day'] == 'Sun') | (data['day'] == 'Sat')][['tip','day']]

# 조건을 정확하게 가져왔는지 확인
a = data[(data['day'] == 'Sun') | (data['day'] == 'Sat')]
a['day'].unique()
a[['tip', 'day']]
```

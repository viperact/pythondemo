### where (numpy)

```python
a = np.array([10, 20, 30, 40, 50])
a
```

```python
# where의 인자는 조건이다.(논리값)
# 조건만 넣을 경우 True의 위치(True의 index값)를 리턴한다
np.where(a > 30) # index값으로 리턴된다.
```

```python
# where는 인자 갯수에 따라서 기능이 달라진다.
# 조건이 True이면 0으로 리턴, False이면 100으로 리턴
np.where(a > 30, 0, 100)
```

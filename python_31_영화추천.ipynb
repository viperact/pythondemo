UserID가 5인 사람과 상관관계가 높은 상위 10명을 뽑을 수 있다.

```python
# sort_values()는 값의 크기순으로 데이터 순서를 정렬한다.
# ascending = False : 내림차순 / ascending = True : 오름차순
# default는 ascending = True로 오름차순이다.
small_test_corr[5].sort_values(ascending = False)[1:11] # 0은 자기자신
```

```python
# 함수 정의를 통한 일반화
# user_id 번호에 따른 상관관계를 불러온다
def nearest_user(user_id, n):
  return small_test_corr.loc[user_id].sort_values(ascending = False)[1:n + 1] # n미만이라 + 1
```

### 193번이 본 영화중에 5번이 보지 않은 영화를 추천할 수 있다.  
### 5번은 봤지만 193번이 안본 영화를 추천할 수 있다. (5번과 193번은 상관관계가 높았으므로 즉, 취향이 비슷한 사람이 본 영화를 추천)

```python
# UserID 193번은 보고, 5번은 보지 않은 영화 목록(MovieID)을 뽑아온다.
# 193번이 본 MovieID의 index - 5번이 본 MovieID의 index
unseen = set(movie_seen(193).index) - set(movie_seen(5).index)
len(unseen)
```

```python
unseen_movie = recommendation_data[recommendation_data.MovieID.isin(unseen)]
unseen_movie
```

```python
# 193번 유저가 5점을 준 영화
unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == 193)]
```

```python
# 추천할 영화 제목을 가져온다.
unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == 193)].merge(movies)
```

1. 193번은 보고 5번은 안본 영화
2. recommendation_data에서 위의 1번 결과에 해당하는 목록을 가져옴 (Movie.index)
3. Rating(평점)이 5이고 193번이 본 영화 목록과 장르를 가져와 연결해준다.

```python
# 함수 정의로 일반화

# user_id는 본사람, other_id는 안본사람
# user_id가 other_id에게 추천
# user_id(193), other_id(5)

def unseen_movie(user_id, other_id):
  unseen = set(movie_seen(user_id).index) - set(movie_seen(other_id).index) # 1 # UserID 193번은 보고, 5번은 보지 않은 영화 목록(MovieID)을 뽑아온다.
  unseen_movie = recommendation_data[recommendation_data.MovieID.isin(unseen)] # 2
  return unseen_movie[(unseen_movie.Rating == 5) & (unseen_movie.UserID == user_id)].merge(movies) # 3 # 추천할 영화 제목을 가져온다.
```

```python
# 193번은 보고 5번은 보지 않은 영화중에 평점이 5인 영화를 5번 유저에게 추천
unseen_movie(193, 5)
```

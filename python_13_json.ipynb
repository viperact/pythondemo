```python
dict = {'10':'1056', 'name':'park', '작업':'프리랜서'}

with open('./data_crowling/dictjson1.txt', 'w') as file:
  file.write(json.dumps(dict))
```

```python
dict = {'10':'1056', 'name':'park', '작업':'프리랜서'}

# 한글이 깨지지않고 정상적으로 출력하게 해줌
with open('./data_crowling/dictjson1.txt', 'w') as file:
  file.write(json.dumps(dict, ensure_ascii = False))
```

```python
import json
json_str = '{"name":"kyeongrok", "age":"32"}';
print(type(json_str)) # str
json_obj = json.loads(json_str)
# json_loads로 가져오면 type이 바뀜
print(json_obj)
print(type(json_obj)) # dict
```

```python
import json
json_str = '[100, 200, 300, 400]'
print(type(json_str)) # str

json_obj = json.loads(json_str)
print(type(json_obj)) # list

print(json_obj)
print(json_obj[0])
print(type(json_obj[0])) # int
```

```python
import json

json_str = '[{"name":"kyeongrok", "age":"32"}, {"name":"bomi", "age":"25"}]'
print(json_str) 
print(type(json_str)) # str

json_obj = json.loads(json_str)
print(json_obj) 
print(type(json_obj)) # list

print(json_obj[0])
print(type(json_obj[0])) # dict
print(json_obj[0]['name'])
```

```python
import json
json_str = '[{"name":"kyeongrok", "age":"32"}, {"name":"bomi", "age":"25"}]'

json_obj = json.loads(json_str)

for element in json_obj:
  print(element)
  print(element['name'], element['age']) #dict로 가져올 수 있음
```
